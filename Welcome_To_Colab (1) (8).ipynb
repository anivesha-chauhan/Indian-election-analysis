```
# Create mapping of party name to most frequent abbreviation

party_map = (
    df_vidhan.groupby("partyname")["partyabbre"]
    .agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
)

# Fill missing abbreviations
df_vidhan["partyabbre"] = df_vidhan.apply(
    lambda row: party_map[row["partyname"]]
    if pd.isna(row["partyabbre"]) and row["partyname"] in party_map
    else row["partyabbre"],
    axis=1
)

# Fill remaining missing values with IND
df_vidhan["partyabbre"] = df_vidhan["partyabbre"].fillna("IND")

print("Missing Party Abbreviations Filled Successfully")
```

```
# Replace inconsistent abbreviations

df_vidhan["partyabbre"] = df_vidhan["partyabbre"].replace({
    "INC(I)": "INC"
})

print("Party Abbreviations Standardized Successfully")
```

```
df_vidhan["year"] = df_vidhan["year"].astype(int)

print("Year Converted to Integer")
```

```
df_vidhan["cand_sex"] = df_vidhan["cand_sex"].fillna("U")

print("Missing Gender Filled")
```

```
df_vidhan = df_vidhan.dropna(subset=["totvotpoll"])

print("Rows with Missing Total Votes Removed")
```

```
# Most frequent assembly constituency name mapping

ac_map = (
    df_vidhan.groupby(["st_name", "ac_no"])["ac_name"]
    .agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
)

# Fill missing ac_name
df_vidhan["ac_name"] = df_vidhan.apply(
    lambda row: ac_map[row["st_name"], row["ac_no"]]
    if pd.isna(row["ac_name"]) and (row["st_name"], row["ac_no"]) in ac_map.index
    else row["ac_name"],
    axis=1
)

print("Missing Assembly Names Filled Successfully")
```

```
# Calculate number of unique candidates per seat each year

avg_candidates = (
    df_vidhan.groupby(["year", "ac_no"])["cand_name"]
    .nunique()
    .groupby("year")
    .mean()
    .reset_index(name="Average_Candidates")
)

avg_candidates.head()
```

```
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.plot(
    avg_candidates["year"],
    avg_candidates["Average_Candidates"],
    marker="o"
)

plt.title("Average Candidates per Seat Over Time")
plt.xlabel("Election Year")
plt.ylabel("Average Candidates per Seat")
plt.grid(True)

plt.show()
```

The line chart shows how the average number of candidates contesting each assembly seat has changed over different election years. An increasing trend indicates higher electoral competition, while a decreasing trend suggests fewer candidates per constituency. The fluctuations across years reflect changes in political participation, party strategies, and the number of independent candidates. Overall, this analysis helps understand how electoral competition has evolved over time.

```
# Calculate voter turnout by year

turnout = (
    df_vidhan.groupby(["year", "ac_no"])
    .agg({
        "totvotpoll": "max",
        "electors": "max"
    })
    .reset_index()
)

turnout = turnout.groupby("year").sum()

turnout["Turnout_Percent"] = (
    turnout["totvotpoll"] / turnout["electors"]
) * 100

turnout.head()
```

```
plt.figure(figsize=(10,5))

plt.plot(
    turnout.index,
    turnout["Turnout_Percent"],
    marker="o"
)

plt.title("Voter Turnout Percentage by Year")
plt.xlabel("Year")
plt.ylabel("Turnout (%)")
plt.grid(True)

plt.show()
```

The graph illustrates the voter turnout percentage across different election years. It highlights whether voter participation has increased, decreased, or fluctuated over time. Higher turnout generally reflects stronger public engagement in the democratic process, while lower turnout may indicate reduced participation due to political, social, or administrative factors. This trend helps evaluate the overall health of electoral participation.

```
# Identify winners

winners = df_vidhan.loc[
    df_vidhan.groupby(["year", "st_name", "ac_no"])["totvotpoll"].idxmax()
]

winners.head()
```

```
gujarat_winners = winners[winners["st_name"] == "Gujarat"]

gujarat_winners.head()
```

Winners were identified by selecting the candidate with the highest number of votes in each constituency for every election year. This method ensures that only the officially elected representatives are included in further analysis. Filtering the data for Gujarat provides a clear picture of electoral outcomes within the state and forms the basis for studying party dominance over time.

```
# Vote Share

vote_share = (
    df_vidhan.groupby("partyabbre")["totvotpoll"]
    .sum()
    .reset_index(name="Party_Votes")
)

vote_share["Vote_Share"] = (
    vote_share["Party_Votes"] /
    vote_share["Party_Votes"].sum()
) * 100

# Seat Share

seat_share = (
    winners["partyabbre"]
    .value_counts(normalize=True)
    .mul(100)
    .reset_index()
)

seat_share.columns = ["partyabbre", "Seat_Share"]

# Merge

comparison = vote_share.merge(
    seat_share,
    on="partyabbre",
    how="left"
).fillna(0)

comparison = comparison.sort_values(
    "Vote_Share",
    ascending=False
).head(10)

comparison
```

```
plt.figure(figsize=(12,6))

comparison.set_index("partyabbre")[["Vote_Share","Seat_Share"]].plot(
    kind="barh"
)

plt.title("Vote Share vs Seat Share")
plt.xlabel("Percentage")

plt.show()
```

This comparison shows whether political parties convert their vote share into seats efficiently. Some parties may secure a high percentage of votes but win fewer seats, while others gain a larger share of seats with a relatively smaller vote share. This difference reflects the impact of India's First-Past-The-Post (FPTP) electoral system, where winning individual constituencies matters more than overall national vote share.

```
# Gujarat Data
gujarat = df_vidhan[df_vidhan["st_name"] == "Gujarat"]

# Total votes by party and year
party_votes = gujarat.groupby(["year", "partyabbre"])["totvotpoll"].sum().reset_index()

# Total votes per year
total_votes = gujarat.groupby("year")["totvotpoll"].sum().reset_index()
total_votes.columns = ["year", "total_votes"]

# Merge
vote_share = pd.merge(party_votes, total_votes, on="year")

# Vote Share %
vote_share["vote_share"] = (vote_share["totvotpoll"] / vote_share["total_votes"]) * 100

# Top 3 Parties
top3 = vote_share.groupby("partyabbre")["totvotpoll"].sum().nlargest(3).index

vote_share = vote_share[vote_share["partyabbre"].isin(top3)]

vote_share.head()
```

```
plt.figure(figsize=(10,6))

for party in top3:
    temp = vote_share[vote_share["partyabbre"] == party]
    plt.plot(temp["year"], temp["vote_share"], marker="o", label=party)

plt.title("Vote Share Trend of Top 3 Parties in Gujarat")
plt.xlabel("Year")
plt.ylabel("Vote Share (%)")
plt.legend()

plt.show()
```

The line graph shows how the vote share of the top three political parties in Gujarat has changed over time. Some parties gained support in certain election years while others experienced a decline. These changes reflect shifts in voter preferences and the changing political landscape of the state.

```
from scipy.stats import f_oneway

groups=[]

for party in top5:
    groups.append(
        vote_share[
            vote_share["partyabbre"]==party
        ]["vote_share_percent"]
    )

F,P = f_oneway(*groups)

print("F Statistic =",F)
print("P Value =",P)
```

```
from scipy.stats import ttest_ind

party1 = top5[0]
party2 = top5[1]

a = vote_share[vote_share.partyabbre==party1]["vote_share_percent"]
b = vote_share[vote_share.partyabbre==party2]["vote_share_percent"]

t,p = ttest_ind(a,b)

print("T Statistic =",t)
print("P Value =",p)
```
